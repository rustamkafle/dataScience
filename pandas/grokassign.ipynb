```python
import pandas as pd
df = pd.read_csv("sales_data.csv")
df.head(5)
```

```python
print(f"there are {df.isna().sum().sum()} missing values")
```

```python
df
```

```python
df['price']=df['price'].fillna(df['price'].median())
```

```python
df['quantity_sold']  = df['quantity_sold'].fillna(0)
df
```

```python
print(f"there are {df.duplicated().sum().sum()} duplicate values")
```

```python
df['review_text'] = df['review_text'].str.lower()
df
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
df
```

```python
df.loc[(df['category']=='Electronics') & (df['price']>500)]
```

```python
df.query("region=='North' and quantity_sold>1")
```

```python
dfA = df.sort_values('price')
```

```python
dfA.reset_index(drop=True, inplace=True)
dfA
```

```python
df.rename(columns={"quantity_sold":"unit_sold"}, inplace=True)
```

```python
df
```

```python
df['Total Revenue'] = df.apply(lambda row: row['price']*row['unit_sold'], axis=1)
df
```

```python
df = df[["customer_id", "product_name", "category", "price", "unit_sold", "Total Revenue", "order_date", "region", "review_text"]]
df
```

```python
df.melt(
    id_vars=['customer_id','product_name'],
    value_vars=['price','unit_sold'],
    var_name='sell_category',
    value_name='unit'

)
```

```python
df.groupby("category")["Total Revenue"].mean()
```

```python
df.groupby("category").agg({'unit_sold':"count", "price":"mean"})
```
