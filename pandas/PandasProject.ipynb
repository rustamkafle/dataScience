## Appending to the dataFrame

```python
new_movies = ['GharJwai','Romance',6.6,2024,88,91,229,'2024-07-16']
df = pd.concat([df, pd.DataFrame([new_movies], columns=df.columns)], ignore_index=True)
```

## Data Cleaning & Preprocessing

```python
df.isna().sum().sum() # there are zero null values in the dataFrame
```

```python
df['Release_Date'] = pd.to_datetime(df['Release_Date'])
df
```

```python
df.duplicated().sum().sum() 
# no duplicates found
```

## Data Transformation

```python
df['Decades'] = (df['Release_Year']//10)*10
df['Rating'] = (df['Rating']*10).astype('int32')
df['Tag'] = df['Rating'].apply(lambda x: "Flop" if x<=80 else "Hit")
df
```

## Melt

```python
df2 = df[['Title','Critic_Rating','Audience_Rating']]
```

```python
df2
```

```python
df3 = df2.melt(
    id_vars=["Title"],
    value_vars=["Critic_Rating","Audience_Rating"],
    var_name="Rating Type",
    value_name="Rate"

)
```

```python
df3
```

## Pivot

```python
df3 = df3.pivot(index="Title",columns=["Rating Type"], values="Rate")
df3.columns.name = None 
df3
```
