```python
import time as t
import numpy as np
import matplotlib.pyplot as plt
def python_list():
    l1 = list(range(10000))
    l2 = list(range(10000))
    result = []
    for i in range(len(l1)):
        result.append(l1[i]*l2[i])
    #print(result)

def numpy_array():
    arr1 = np.arange(10000)
    arr2 = np.arange(10000)
    result = arr1 * arr2
    #print(result)
```

```python
def list_operation_time(iteration_number):
    for i in range(iteration_number):
        begin_list = t.time()
        python_list()
        end_list = t.time()
        operation_time_list.append((end_list-begin_list)*1000)
```

```python
def numpy_operation_time(iteration_number):
    for i in range(iteration_number):
        begin_numpy = t.time()
        numpy_array()
        end_numpy = t.time()
        operation_time_numpy.append((end_numpy-begin_numpy)*1000)
```

```python
operation_time_list = []
operation_time_numpy = []
iteration = 1000
list_operation_time(iteration)
numpy_operation_time(iteration)

x = np.arange(1, iteration+1)
plt.plot(x, operation_time_list, label="Python List")
plt.plot(x, operation_time_numpy, label="Numpy Array")
plt.xlabel("Operation Number")
plt.ylabel("Time Taken")
plt.legend()
plt.show()
```

```python
import statistics as st
avg_list_time = st.mean(operation_time_list)
avg_numpy_time = st.mean(operation_time_numpy)
print("Mean Time for Python List", avg_list_time, " milliseconds")
print("Mean Time for Numpy Array", avg_numpy_time, " milliseconds")
print()
print(f"In average,\nNumpy Array operation is {int(avg_list_time/avg_numpy_time)} times faster than Python list operation")
```
