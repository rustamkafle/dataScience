```python
import matplotlib.pyplot as plt
```

## Plotting Simple Bar Graph

```python
plt.bar([1,2,3],[3,2,5])
plt.show()
```

## Using Keys Values for x-axis and y-axis respectively

```python
dict = {'A': 40, 'B': 70, 'C': 30, 'D': 85}
X = dict.keys()
Y = dict.values()
plt.bar(X,Y)

plt.xlabel("Student Name")
plt.ylabel("Test Score")
plt.title("Student Test Scores")

plt.show()
```

## Using xticks() and yticks()
#### we may not like how each of those bars are represented with alphabets. We want to replace them with what we want. In such cases, we make use of xtick. In y-axis, you may not wish too see all those 10 20 30 upto 80, but just those significant ones. We can also achieve that using tick. YUP!

```python
dict = {'A': 40, 'B': 70, 'C': 30, 'D': 85}
new_label = ['Ram', 'Shyam', 'Hari', 'Sita']
X = list(dict.keys())
Y = list(dict.values())
plt.bar(X,Y)

plt.xlabel("Student Name")
plt.ylabel("Test Score")

plt.xticks(X, new_label, rotation=45, ha='right') # plt.xticks(old_label, new_label, otherparameters)
plt.yticks([0,30,60,80]) # just shows ticks for these four values on y-axis
plt.title("Student Test Scores")

plt.show()
```

## Changing width of bars, alignment, and color

```python
width_list = [0.25,0.75]

plt.bar([1,3],[3,5], width=width_list, align='edge', color='violet')

plt.xticks([1,3],['Rustam','Richa'], rotation=30)
plt.yticks([3,5],['pudke','aglo'])

plt.show()
```

# Horizontal bar chart

```python
width_list = [0.25,0.75]

plt.barh([1,3],[3,5], width_list, color='violet')

plt.yticks([1,3],['Rustam','Richa'], rotation=30)
plt.xticks([3,5],['pudke','aglo'])

plt.gca().invert_yaxis()

plt.show()
```

In plt.bar(), you typically use the keyword argument width= to specify the widths of your vertical bars.

In plt.barh(), you provide the list of widths (lengths) as the second positional argument, right after the y-coordinates.

# Bar with Error

```python
width_list = [0.25,0.75]
error = [0.5, 2]
plt.barh([1,3],[3,5], width_list, xerr=error, color='violet')

plt.yticks([1,3],['Rustam','Richa'], rotation=30)
plt.xticks([3,5],['pudke','aglo'])

plt.gca().invert_yaxis()

plt.show()
```

# Coloring

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['A', 'B', 'C', 'D']
values = [25, 40, 30, 55]
errors = [2, 3, 1.5, 4]

plt.bar(categories, values, yerr=errors, capsize=7, color=['skyblue','green','pink','red'], edgecolor='black')

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart with Error Bars")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['A', 'B', 'C', 'D']
values = [25, 40, 30, 55]
errors = [2, 3, 1.5, 4]

plt.bar(categories, values, yerr=errors, color='w', edgecolor='blue',linewidth=[10,5,2,1] )
# use linewidth to control the border width of bar and edgecolor to fill the border
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart with Error Bars")
plt.show()
```
