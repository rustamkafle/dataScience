## Decorating our Plot

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
plt.plot(years, runs) # x-axis, y-axis
plt.xlabel('Years')
plt.ylabel('Runs')
plt.title('Player\'s Stats')
plt.grid(True)
plt.show()
```

```python
cubic = poly1d([1,0,0,0])
x1 = np.arange(-2,3)
```

```python
plt.xkcd()
plt.plot(x1,cubic(x1), label="cubic")
plt.plot(x1, np.sin(2 * np.pi * x1), color='green', label="sin$")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```
