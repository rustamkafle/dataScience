```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(1,10,9999)
y = np.exp(x)
z = np.log(x)
a = np.sin(x)

plt.figure(figsize=(12,4))

plt.suptitle("Three Important Graphs", fontsize=16, color="#2bb83f")

plt.subplot(1,3,1)
plt.plot(x, y, color="indigo")
plt.title("Exponential")
plt.grid(True)

plt.subplot(1,3,2)
plt.plot(x, z, color="red")
plt.title("Logarithm")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.subplot(1,3,3)
plt.plot(x, a, color="black")
plt.title("sinusoidal")
plt.grid(True)

plt.tight_layout() # to prevent overlaps
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10,10,9999)
y = np.exp(x)
z = np.arctan(x)
a = np.sin(x)
b = x**2

plt.figure(figsize=(12,4))
plt.suptitle("Four Important Graphs", fontsize=16, color="#2bb83f")

plt.subplot(2,2,1)
plt.plot(x, y, color="green")
plt.title("Exponential")

plt.subplot(2,2,2)
plt.plot(x, z, color="red")
plt.title("Arc Tan")

plt.subplot(2,2,3)
plt.plot(x, a, color="black")
plt.title("sinusoidal")

plt.subplot(2,2,4)
plt.plot(x, b, color="#0000FF")
plt.title("Parabola")

plt.tight_layout() # to prevent overlaps
```

```python
fig, axs = plt.subplots(1,2, figsize=(10,5))

# uses of axs
axs[0].plot(x, y, label="sth", color="green")
axs[0].set_ylabel("y-axis")
axs[0].set_title("Exp")
axs[0].legend()

axs[1].plot(x, z)
axs[1].set_xlabel("x-axis")
axs[1].set_title("arc tan")
axs[1].grid(True)
axs[1].axis("off")

# uses of fig
fig.suptitle("Helllllooo Bachhhooooo", fontsize=15)
fig.tight_layout()
fig.subplots_adjust(hspace=2, wspace=0.5)
fig.savefig("my_figure.png")
```

```python
fig, axs = plt.subplots(2,2, figsize=(10,5))

axs[0,0].stackplot([1,2,3],[4,5,6],[5,6,7],[5,3,1])
axs[0,0].set_title("Testing-1")

axs[0,1].scatter([1,2,3],[-4,-5,-6])
axs[0,1].set_title("Testing-2")

axs[1,0].bar([1,2,3],[1,4,6])
axs[1,0].set_title("Testing-3")

axs[1,1].pie([1,2,3])
axs[1,1].set_title("Testing-4")

fig.tight_layout()
fig.suptitle("Matplotlib", fontsize=15, color="green")
```
