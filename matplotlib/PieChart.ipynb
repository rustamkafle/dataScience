```python
import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight")
```

## Simplest Pie Chart

```python
plt.pie([8, 8, 4, 2, 2])
plt.show()
```

## Labelling our PieChart

```python
activities = ['Sleep', 'Work/Study', 'Leisure/Entertainment', 'Exercise', 'Eating']
hours = [8, 8, 4, 2, 2]
plt.pie(hours, labels=activities)
plt.show()
```

## Giving our custom colors to each of the pie

```python
activities = ['Sleep', 'Work/Study', 'Leisure/Entertainment', 'Exercise', 'Eating']
hours = [8, 8, 4, 2, 2]
colors = ['#b88e2b', '#2bb840', '#2ba2b8', '#642bb8', '#b82b78']
plt.pie(hours, labels=activities, colors=colors)
plt.show()
```

## Custom Starting Angle

```python
activities = ['Sleep', 'Work/Study', 'Leisure/Entertainment', 'Exercise', 'Eating']
hours = [8, 8, 4, 2, 2]
colors = ['#b88e2b', '#2bb840', '#2ba2b8', '#642bb8', '#b82b78']
plt.pie(hours, labels=activities, colors=colors, startangle=90) # The pie for the first item in the list, starts at angle 90 degree
plt.show()
```

## Displaying Percentage of each Pie

```python
activities = ['Sleep', 'Work/Study', 'Leisure/Entertainment', 'Exercise', 'Eating']
hours = [8, 8, 4, 2, 2]
colors = ['#b88e2b', '#2bb840', '#2ba2b8', '#642bb8', '#b82b78']
plt.pie(hours, labels=activities, colors=colors, startangle=90, autopct="%1.1f%%")
plt.show()
```

## Taking a slice out of the pie using explode

```python
activities = ['Sleep', 'Work/Study', 'Leisure/Entertainment', 'Exercise', 'Eating']
hours = [8, 8, 4, 2, 2]
colors = ['#b88e2b', '#2bb840', '#2ba2b8', '#642bb8', '#b82b78']
explode=[0,0,0.2,0,0]
plt.pie(hours, labels=activities, colors=colors, startangle=90, autopct="%1.1f%%", explode=explode)
plt.show()
```

## Making slices realistic using shadows

```python
activities = ['Sleep', 'Work/Study', 'Leisure/Entertainment', 'Exercise', 'Eating']
hours = [8, 8, 4, 2, 2]
colors = ['#b88e2b', '#2bb840', '#2ba2b8', '#642bb8', '#b82b78']
explode=[0,0,0.2,0,0]
plt.pie(hours, labels=activities, colors=colors, startangle=90, autopct="%1.1f%%", explode=explode, shadow=True)
plt.show()
```

## Exploring Wedgeprops

```python
activities = ['Sleep', 'Work/Study', 'Leisure/Entertainment', 'Exercise', 'Eating']
hours = [8, 8, 4, 2, 2]
explode=[0,0,0.2,0,0]

properties = {'edgecolor':'black', 'linewidth':3, "linestyle":"-.", 'fill':False}
plt.pie(hours, labels=activities, autopct="%1.1f%%", explode=explode, wedgeprops=properties)
plt.show()
```

## Ultimate Adornment: Hatch Slices

```python
activities = ['Sleep', 'Work/Study', 'Leisure/Entertainment', 'Exercise', 'Eating']
hours = [8, 8, 4, 2, 2]
explode=[0,0,0.2,0,0]
patterns = ['**O', 'oO', 'O.O', '.||.', '++']

plt.pie(hours, labels=activities, explode=explode, hatch=patterns) # you can look up the documentation to see different hatch patterns
plt.show()
```
