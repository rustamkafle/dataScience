```python
import matplotlib.pyplot as plt
import numpy as np
```

## Using bins as integer

```python
data = np.random.randint(0,101,450)
plt.hist(data, bins=5, edgecolor="black")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

## Using custom bins as list

```python
ages = np.random.randint(10,71,100)
bins= [10,20,40,60,70]

plt.hist(ages,bins=bins,edgecolor="black")
plt.title("Age Distribution of YouTube Viewers")
plt.xlabel("Age Group")
plt.ylabel("Number of Viewers")
#plt.grid(True, linestyle="--",alpha=0.5)
plt.show()
```

## Using axvline

```python
ages = np.random.randint(10,71,100)

plt.axvline(35,color="red",linestyle="--",linewidth=2,label="hellooooooooo")
plt.hist(ages,bins=6,edgecolor="black")
plt.title("Age Distribution of YouTube Viewers")
plt.xlabel("Age Group")
plt.ylabel("Number of Viewers")
#plt.grid(True, linestyle="--",alpha=0.5)
plt.legend()
plt.show()
```
