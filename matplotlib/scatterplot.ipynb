```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
study_hours = np.arange(1,10)
exam_scores= [40,90,50,55,60,25,70,75,85]
```

```python
plt.scatter(study_hours, exam_scores)
plt.grid(True)
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.title("Study vs Marks")
plt.show()
```

## Adding color and size

```python
colors = ['red' if score<70 else 'green' for score in exam_scores]
sizes = [score*2 for score in exam_scores]
plt.scatter(study_hours, exam_scores, c=colors, s=sizes)
plt.grid(True)
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.title("Study vs Marks")
plt.show()
```

## Colour Map CMAP - comparing study hours, exam scores and EQ (3 parameters)

```python
#colors = ['red' if score<70 else 'green' for score in exam_scores]
sizes = [score*2 for score in exam_scores]
emotional_intelligence = [10,59,56,54,10,18,22,17,19]
plt.scatter(study_hours, exam_scores, c=emotional_intelligence, s=sizes, cmap="Greens")
plt.colorbar(label="Score")
plt.grid(True)
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.title("Study vs Marks")
plt.show()
```

## Adding Annotations

```python
sizes = [score*2 for score in exam_scores]
emotional_intelligence = [30,59,56,54,10,18,22,17,19]
plt.scatter(study_hours, exam_scores, c=emotional_intelligence, s=sizes, cmap="binary")
plt.colorbar(label="Emotional Intelligence")
# Adding labels
for i in range(len(study_hours)):
    plt.annotate(f"S{i+1}", (study_hours[i]+0.2, exam_scores[i]))

plt.title("Scatter Plot with Annotations")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.grid(True)
plt.show()
```

## Multiple Groups in one plot

```python
# Assume two groups: Class A and Class B
plt.style.use("fast")
class_a_hours = [2, 4, 6, 8]
class_a_scores = [45, 55, 65, 85]

class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 50, 60, 70, 90]

plt.scatter(class_a_hours, class_a_scores, edgecolor='green', linewidth=2, alpha=0.2, label='Class A', color='brown')
plt.scatter(class_b_hours, class_b_scores, edgecolor='blue', label='Class B', color='orange')

plt.title('Scatter Plot: Class A vs Class B')

plt.xlabel('Study Hours')
plt.xscale("log")

plt.ylabel('Exam Score')
plt.yscale("log")

plt.legend()
plt.show()
```
