```python
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
sns.set_theme(style="dark")
x = np.linspace(1,10,9999)
y = np.sin(x)
z = np.cos(x)
sns.lineplot(x=x, y=y, color="brown")
sns.lineplot(x=x, y=z, color="orange")
sns.lineplot(x=x, y=0)
plt.title("Beauiful sin cos")
plt.show()
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.lineplot(x="total_bill", y="tip", data=tips)
plt.title("Line Plot", color="green")
plt.axis("off")
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Scatter Plot", color="green")
```

```python
sns.barplot(x="total_bill", y="tip", data=tips)
plt.title("Bar Plot", color="green")
plt.xlim(10,12)
```

```python
sns.boxplot(x="total_bill", y="tip", data=tips)
plt.title("Box Plot", color="green")
plt.xlim(10,12)
```
